# Python Caching Basics

## Learning Objectives

- Use lru_cache for simple caching
- Understand eviction and invalidation
- Decide when caching is appropriate

---

## 1. lru_cache Example

```python
from functools import lru_cache

@lru_cache(maxsize=128)
def compute(x: int) -> int:
    return x * x

print(compute(4))
print(compute(4))  # cached
```

## Summary

- Cache expensive computations
- Use maxsize to limit memory
- Plan invalidation for stale data
