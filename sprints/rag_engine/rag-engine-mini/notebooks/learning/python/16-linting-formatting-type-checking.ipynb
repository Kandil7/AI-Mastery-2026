## 1. Formatting with Black

```python
commands = [
    'black --check .',
    'black .',
]
for cmd in commands:
    print(cmd)
```

## 2. isort

```python
commands = [
    'isort --check-only .',
    'isort .',
]
for cmd in commands:
    print(cmd)
```

## 3. flake8

```python
print('flake8 src tests')
```
