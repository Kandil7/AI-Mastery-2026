# Python Documentation and Docstrings

## Learning Objectives

- Write clear function docstrings
- Document classes and modules
- Keep docs concise and accurate

---

## 1. Function Docstring

```python
def add(a: int, b: int) -> int:
    """Add two integers.

    Args:
        a: First integer.
        b: Second integer.

    Returns:
        Sum of a and b.
    """
    return a + b

print(add(2, 3))
```

## 2. Class Docstring

```python
class DocumentService:
    """Service for document operations.

    Handles validation and persistence logic.
    """

    def __init__(self, name: str) -> None:
        self.name = name

print(DocumentService('docs').name)
```

## Summary

- Use docstrings for public functions and classes
- Keep docs updated
- Avoid redundant explanations
