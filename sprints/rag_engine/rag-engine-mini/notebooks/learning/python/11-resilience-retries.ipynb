## 2. Retry with Backoff

```python
import time

def retry_backoff(operation, attempts: int = 3, base_delay: float = 0.5):
    last_error = None
    for attempt in range(attempts):
        try:
            return operation()
        except Exception as exc:
            last_error = exc
            time.sleep(base_delay * (2 ** attempt))
    raise last_error

print('retry_backoff ready')
```
