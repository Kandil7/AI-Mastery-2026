# Python Automation with Make and Scripts

## Learning Objectives

- Use Make targets for repeatable workflows
- Understand why automation matters
- Keep dev commands consistent

---

## 1. Common Targets

```python
targets = [
    'make install',
    'make test',
    'make lint',
    'make format',
    'make run',
]
for target in targets:
    print(target)
```

## 2. Example Makefile Target

```python
example = [
    '.PHONY: test',
    'test:',
    '	pytest tests/ -v',
]
for line in example:
    print(line)
```

## Summary

- Make targets standardize workflows
- Scripts can wrap multi-step tasks
- Keep commands documented and consistent
