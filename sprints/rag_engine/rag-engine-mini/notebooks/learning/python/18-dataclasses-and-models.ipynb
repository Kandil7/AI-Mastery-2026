# Python Data Classes and Models

## Learning Objectives

- Create dataclasses for domain entities
- Compare dataclasses and Pydantic models
- Use frozen dataclasses for immutability

---

## 1. Basic Dataclass

```python
from dataclasses import dataclass

@dataclass
class Document:
    id: str
    title: str
    content: str

doc = Document(id='1', title='Doc', content='Text')
print(doc)
```

## 2. Frozen Dataclass

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class User:
    id: str
    name: str

user = User(id='1', name='Alice')
print(user)
```

## Summary

- Dataclasses are lightweight and clear
- Pydantic is for validation at boundaries
- Use frozen dataclasses for immutability
