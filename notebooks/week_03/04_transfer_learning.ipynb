## The Power of Transfer

### From Scratch
- Train: 1M images, 1 week, $10K compute
- Accuracy: 70%

### With Transfer
- Fine-tune: 10K images, 1 hour, $10 compute  
- Accuracy: 85%+

**Why it works**: Low-level features (edges, textures) transfer!

## Common Strategies

### 1. Feature Extraction
- **Freeze** all layers except last
- **Train** only final classifier
- **Use when**: Small dataset, similar domain

### 2. Fine-Tuning
- **Unfreeze** some layers
- **Train** with small learning rate
- **Use when**: Medium dataset, related domain

### 3. Full Fine-Tuning
- **Unfreeze** all layers
- **Train** entire network
- **Use when**: Large dataset, different domain

## Transfer Learning Recipe

1. **Load pre-trained model**
2. **Replace final layer** (match your classes)
3. **Freeze early layers**
4. **Train on your data**
5. **Gradually unfreeze** and fine-tune
