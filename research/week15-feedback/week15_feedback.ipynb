# Section 1: Feedback Collection

```python
feedback_schema = '''
// Feedback data model
{
  "response_id": "uuid",
  "query": "user question",
  "response": "LLM answer",
  "rating": 1-5,
  "feedback_type": "helpful|unhelpful|incorrect|offensive",
  "comment": "optional text",
  "timestamp": "2024-01-01T00:00:00Z"
}
'''
print(feedback_schema)
```

```python
feedback_api = '''
@app.post("/feedback")
async def submit_feedback(feedback: FeedbackInput):
    # Store in database
    await db.feedback.insert(feedback.dict())
    
    # Trigger analysis if negative
    if feedback.rating < 3:
        await enqueue_for_review(feedback)
    
    return {"status": "received"}
'''
print(feedback_api)
```
