## 2. Coding Interview Tips

### 2.1 The UMPIRE Framework

| Step | Action | Time |
|------|--------|------|
| **U**nderstand | Clarify requirements, edge cases | 5 min |
| **M**atch | Identify patterns, similar problems | 2 min |
| **P**lan | Outline approach, discuss trade-offs | 5 min |
| **I**mplement | Write clean, working code | 20 min |
| **R**eview | Test with examples, debug | 5 min |
| **E**valuate | Analyze complexity, optimize | 5 min |

````python
# Example: Common ML coding problem - k-means
import numpy as np

def kmeans(X, k, max_iters=100):
    """
    K-means clustering implementation.
    
    Args:
        X: Data points (n_samples, n_features)
        k: Number of clusters
        max_iters: Maximum iterations
    
    Returns:
        labels: Cluster assignments
        centroids: Cluster centers
    """
    n_samples, n_features = X.shape
    
    # Step 1: Initialize centroids randomly
    indices = np.random.choice(n_samples, k, replace=False)
    centroids = X[indices].copy()
    
    for _ in range(max_iters):
        # Step 2: Assign points to nearest centroid
        distances = np.array([
            np.sqrt(np.sum((X - c) ** 2, axis=1))
            for c in centroids
        ])
        labels = np.argmin(distances, axis=0)
        
        # Step 3: Update centroids
        new_centroids = np.array([
            X[labels == i].mean(axis=0) if np.sum(labels == i) > 0 else centroids[i]
            for i in range(k)
        ])
        
        # Step 4: Check convergence
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
    
    return labels, centroids

# Test
np.random.seed(42)
X = np.vstack([np.random.randn(30, 2) + [i*3, i*3] for i in range(3)])
labels, centroids = kmeans(X, k=3)
print(f"Cluster sizes: {[np.sum(labels == i) for i in range(3)]}")
````

## 4. ML Theory Review

### 4.1 Must-Know Concepts

| Topic | Key Concepts |
|-------|-------------|
| **Supervised Learning** | Regression, classification, loss functions |
| **Deep Learning** | Backprop, optimizers, regularization |
| **Transformers** | Attention, BERT, GPT, fine-tuning |
| **Evaluation** | Precision/recall, AUC, cross-validation |
| **RAG** | Retrieval, reranking, generation |

````python
# Quick concept review: Attention mechanism

def attention(Q, K, V):
    """
    Scaled dot-product attention.
    
    Attention(Q, K, V) = softmax(QK^T / âˆšd_k) V
    
    Interview tip: Be ready to explain:
    - Why scale by âˆšd_k? (prevent softmax saturation)
    - What are Q, K, V? (queries match keys to retrieve values)
    - Why is attention powerful? (captures long-range dependencies)
    """
    d_k = K.shape[-1]
    scores = Q @ K.T / np.sqrt(d_k)
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)
    return weights @ V

print("Attention formula: softmax(QK^T / âˆšd_k) V")
````
