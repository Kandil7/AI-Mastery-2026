# Python CLI Tools with argparse

## Learning Objectives

- Define CLI arguments and flags
- Parse inputs safely
- Structure a simple CLI entry point

---

## 1. Basic CLI

```python
import argparse

parser = argparse.ArgumentParser(description='Example CLI')
parser.add_argument('name', help='Name to greet')

# args = parser.parse_args()
print('argparse example ready')
```

## 2. Optional Flags

```python
import argparse

parser = argparse.ArgumentParser()
parser.add_argument('--limit', type=int, default=10)

print('flags example ready')
```

## 3. Subcommands

```python
import argparse

parser = argparse.ArgumentParser()
sub = parser.add_subparsers(dest='command', required=True)
hello = sub.add_parser('hello')
hello.add_argument('name')

print('subcommands example ready')
```

## Summary

- Use argparse for CLI parsing
- Keep CLI thin and logic in modules
- Use subcommands for multi-action tools
