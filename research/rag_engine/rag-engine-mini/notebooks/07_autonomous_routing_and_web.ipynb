## 1. Intent Classification

## 2. Privacy Redaction

We MUST protect user data before it ever hits an external API.

```python
sensitive_question = "Can you contact john.doe@email.com at +1-555-0199?"

redacted = privacy.redact(sensitive_question)
print(f"Original: {sensitive_question}")
print(f"Sent to LLM: {redacted}")

# Restore
restored = privacy.restore(f"The assistant will try to reach <EMAIL_0> or <PHONE_0> soon.")
print(f"\nFinal Result for User: {restored}")
privacy.clear()
```
